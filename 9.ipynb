```python
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd

data = sns.load_dataset('titanic')

sns.countplot( x ='sex', data = data)
plt.show()

sns.boxplot(x='sex',y='age', data = data, hue='survived', color = "red")
plt.title("Distribution of age and gender wrt survival rate")
plt.show()
```
