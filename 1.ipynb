```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np 
```

```python
df = pd.read_csv('Datasets/autodata.csv')
```

```python
df.sample(5)
```

```python
print("---Information--\n")
df.info()
```

```python
print("-----Describe the Dataframe-----")
df.describe()
```

```python
print(df.isnull().sum()) 
```

```python
df['stroke'].fillna(method='ffill', inplace=True)
df['horsepower'].fillna(method='ffill', inplace=True)
df['peak-rpm'].fillna(method='ffill', inplace=True)
df['horsepower-binned'].fillna(method='ffill', inplace=True)
```

```python
print(df.isnull().sum()) 
```

```python
print("===Data Standardization====\n")


#Converting miles per galon to liters per 100km

print("-----Standardizing 'city-mpg' column to 'city-L/100km'-----")
df["city-L/100km"] = 235/df["city-mpg"]
df.head()
```

```python
print("======Data Normalization====\n")
print("-------Normalizing 'length', 'width', 'height' column----\n")

df['length']=df['length']/df['length'].max()
df['width']=df['width']/df['width'].max()
df['height']=df['height']/df['height'].max()
print(df[['length','width','height']].head())
```

```python
print("-------turning categorical values into quantitative (numeric) variables---------")
```

```python
df['aspiration'].unique()
```

```python
df['aspiration'].replace('std', 0, inplace=True)
df['aspiration'].replace('turbo', 1, inplace=True)
```
